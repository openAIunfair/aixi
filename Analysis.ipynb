```python
env = "CoinFlip"  # env stands for environment

d = {}
with open(env + '.txt', 'r') as txt:
    txt_lines = txt.read()[:-1].split('\n')
    index = 0
    for line in txt_lines:
        d[2**index] = float(line)
        index += 1
    
print(len(d))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Number of points to plot
points = len(d)

x = [2 ** i for i in list(range(1, len(d), len(d) // points))]
y = [d[i] for i in x]

# Enlarge the figure
plt.figure(figsize=(10,5))

plt.xscale('log')
plt.plot(x, y)
plt.title('Learning Scalability of ' + env)

# Add the line for optimal value as a reference
optimal = [0.7] * len(x)
plt.plot(x, optimal)

plt.xlabel("Experience (Cycles)")

# ticks = list(range(1, len(d), len(d) // 20))
ticks = [10 ** i for i in range(6)]
# plt.xticks([2**i for i in ticks], ticks)
plt.xticks(ticks, ticks)

plt.ylabel("Average Reward per Cycle")
plt.plot(x, y,'-',label="MC-AIXI")
plt.plot(x, optimal,'-',color='r',label="optimal")
plt.legend(loc='lower right')
plt.show()
```

```python
env = "cheese_maze"  # env stands for environment

d = {}
with open(env + '.csv', 'r') as csv:
    csv_lines = csv.read()[:-1].split('\n')
    index = 0
    for line in csv_lines:
        l = line.split(", ")
        d[int(l[0])] = float(l[7])
        index += 1
    
print(len(d))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Number of points to plot
points = len(d)

x = list(range(1, len(d), len(d) // points))
y = [d[i] for i in x]

# Enlarge the figure
plt.figure(figsize=(10,5))

# plt.xscale('log')
plt.plot(x, y)
plt.title('Learning Scalability of ' + env)

# Add the line for optimal value as a reference
optimal = [38 / 3] * len(x)
plt.plot(x, optimal)

plt.xlabel("Experience (Cycles)")

ticks = list(range(1, len(d), len(d) // 20))
# plt.xticks(ticks, [i for i in ticks])
plt.xticks(ticks, ticks)

plt.ylabel("Average Reward per Cycle")
plt.plot(x, y,'-',label="MC-AIXI")
plt.plot(x, optimal,'-',color='r',label="optimal")
plt.legend(loc='lower right')
plt.show()
```
